How many people of each race are represented in this dataset?

```python
racename_count = df['race'].value_counts()
racename_count
```

What is the average age of men?

```python
avg_age_men=df.loc[df['sex'] == 'Male', 'age'].mean().round(1)
print("avg_age_men:",avg_age_men)
```

What is the percentage of people who have a Bachelor's degree?

```python
batchelor_degree = round((df.loc[df['education'] == 'Bachelors', 'education'].size * 100 / df.shape[0]), 1)
print("% of  Bachelor's degree:" , batchelor_degree)

```

What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?

```python
advance_education = df[(df['education'] == 'Bachelors') | (df['education'] == 'Masters') | (df['education'] == 'Doctorate')]
advance_and_rich = round((advance_education[advance_education['salary'] == '>50K'].shape[0] * 100 / advance_education.shape[0]), 1)
print("% of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K:", advance_and_rich)
```

What percentage of people without advanced education make more than 50K?

```python
without_degree = df[(df['education'] != 'Bachelors') & (df['education'] != 'Masters') & (df['education'] != 'Doctorate')]
without_degree_rich = round((without_degree[without_degree['salary'] == '>50K'].shape[0] * 100 / without_degree.shape[0]), 1)
print("percentage of people without advanced education make more than 50K:", without_degree_rich)
```

What is the minimum number of hours a person works per week?

```python
min_hours = df['hours-per-week'].min()
print("minimum number of hours a person works per week:", min_hours)
```

What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?

```python
num_min_workers = df[df['hours-per-week'] == min_hours].shape[0]
num_min_rich = df[(df['hours-per-week'] == min_hours) & (df['salary'] == '>50K')].shape[0]
min_rich_percentage = round((num_min_rich * 100 / num_min_workers), 1)
print("percentage of the people who work the minimum number of hours per week have a salary of more than 50K:", min_rich_percentage)
```

What country has the highest percentage of people that earn >50K and what is that percentage?

```python
country_counts = df[df['native-country'] != '?']['native-country'].value_counts()
rich_country_counts = df[(df['salary'] == '>50K') & (df['native-country'] != '?')]['native-country'].value_counts()
earning_percentage = (rich_country_counts / country_counts) * 100
highest_earning_country = earning_percentage.idxmax()
highest_earning_country_percentage = round(earning_percentage.max(), 1)

print("country with the highest percentage of people that earn >50K:", highest_earning_country)
print("percentage:", highest_earning_country_percentage)
```

Identify the most popular occupation for those who earn >50K in India.

```python

India_top_occupation = df.loc[(df['salary'] == '>50K') & (df['native-country'] == 'India'), 'occupation'].value_counts()
print("popular occupations for those who earn >50K in India:\n", India_top_occupation)
India_top_occupation = India_top_occupation.idxmax()
print("\nmost popular occupation for those who earn >50K in India:", India_top_occupation)
```
